## Introduction: 
This project is an implementation of the paper "Generative Poisoning Attack Method Against Neural Networks" by Chaofei Yang, Qing Wu, Hai Li, Yiran Chen.
https://arxiv.org/abs/1703.01340

The goal of this poisoning attack is to generate a poisoning image that is carefully modified so that when it is added into the training set, the average loss of all the training data would increase and decrease the overall performance of the model. This attack is a white box attack which means the adversaries have the knowlege of the model structure and they are able to inject data into the training set. This paper introduced two methods. They are the direct gradient method and generative method. The direct gradient method is very time consuming, and it is not practical to implement on large datasets. Therefore I made a little modification. With this modification, the poisoning image may not "kill" the entire model, but it is guaranteed to cause the target image to be miss classified.


```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import time
import os
import copy
from torch.autograd import Variable
import torch.nn.functional as F
from torchvision.utils import save_image
from torch.autograd import grad

import higher

from torchvision.utils import save_image

from tqdm import tqdm_notebook as tqdm
```

## Direct Gradient method:
The idea of the direct gradient method is to maximize the loss calculated by feeding the target image into the poisoned model. The following code shows how to implement this algorithm step by step with detailed explanations

### Step 1: Train a neural network with normal(untainted) data.
The very first step is to train a neural net with untainted data. In this project, I am training a resnet18 from scratch. The dataset consists of animal images. The animal dataset has four classes, and they are butterflies, chickens, elephants, and horses.

```python
#Define data transform
data_transform = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])
    
# Load training data
trainset_dir = './animal_data/train'
trainset = datasets.ImageFolder(trainset_dir, transform=data_transform)
trainset_loader = torch.utils.data.DataLoader(trainset, batch_size = 16, shuffle = True, num_workers = 12)

# Define helper parameters
dataset_sizes = len(trainset)
class_names = trainset.classes
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("Device is set to: ", device)
```

Data visualization to make sure that they are correctly loaded and transformed

```python
# Data visualization
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(trainset_loader))
inputs = inputs[0:4]
classes = classes[0:4]
# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

Start the training process. Here are some details of the training process.

loss_function: CrossEntropyLoss

optimizer: SGD with initial lr set to 0.001 and momentum set to 0.9

learn_rate_scheduler: StepLR with step size of 100 and gamma of 0.1 

```python
train = False
# Train a resnet18 from scratch
resnet = models.resnet18(pretrained = False)

# Note that the resnet18 downloaded from pytroch is trained on 1000 classes. 
# Therefore the fc layer has an output of 1000. We have to manually change the output to 4
resnet_in_feature = resnet.fc.in_features
resnet.fc = nn.Linear(resnet_in_feature, len(class_names))
resnet = resnet.to(device)



criterion = nn.CrossEntropyLoss()
optimizer_ft = optim.SGD(resnet.parameters(), lr=0.001, momentum=0.9)
exp_lr_scheduler = lr_scheduler.StepLR(optimizer_ft, step_size=100, gamma=0.1)

epoch_num = 15
if train:
    
    for epoch in tqdm(range(epoch_num)):
        running_loss = 0.0
        running_corrects = 0

        for inputs, labels in trainset_loader:
            inputs = inputs.to(device)
            labels = labels.to(device)

            optimizer_ft.zero_grad()

            outputs = resnet(inputs)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)

            loss.backward()
            optimizer_ft.step()
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / dataset_sizes
        epoch_acc = running_corrects.double() / dataset_sizes
        print('{} Loss: {:.4f} Acc: {:.4f}'.format(
                    epoch, epoch_loss, epoch_acc))
```

Testing process:

```python
resnet = models.resnet18(pretrained=False)
resnet.fc = nn.Linear(resnet.fc.in_features, 4)
resnet.load_state_dict(torch.load('./poinson_o_dict.pth'))
resnet = resnet.to(device)
# Loading testset
test_data_transform = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])

test_folder_path = './animal_data/test'
test_dataset = datasets.ImageFolder(test_folder_path, transform=test_data_transform)
test_dataloader = torch.utils.data.DataLoader(test_dataset, batch_size = 4, shuffle = True, num_workers = 4)

running_corrects = 0

#test the model
for images, labels in test_dataloader:
    images = images.to(device)
    labels = labels.to(device)
    
    outputs = resnet(images)
    _, pred = torch.max(outputs, 1)
    
    running_corrects += torch.sum(pred == labels)
    
print("Tested {} images".format(len(test_dataset)))
print("{} images are labeled correctly".format(running_corrects))
print("Model accuarycy is: {:.2f}".format(running_corrects.double()/len(test_dataset)))
```

```python
# Save the original (untainted model)
model_o_dict = resnet.state_dict()
torch.save(model_o_dict, './poinson_o_dict.pth')
```

### Step 2: Select a poison image and a target image
The paper suggested that a poison image can either be generated randomly from a uniform distribution or selecting a random image from the training set and set the label to an incorrect one. In this implementation I am selecting a random image from the training set as my poison image

```python
def single_imshow(image, title):
    #Takes in a numpy
    img = image[0].transpose((1,2,0))
    
    img = (img - np.min(img))/(np.max(img)-np.min(img))
    plt.imshow(img)
    plt.title(title)
    plt.show()
# Select two images from the training set and use one of them as poison image and one as target image
poison_dataloader = torch.utils.data.DataLoader(trainset, batch_size=1, shuffle=True,num_workers=4)
target_dataloader = torch.utils.data.DataLoader(trainset, batch_size = 1, shuffle=True, num_workers=4)

poison_image, poison_label = next(iter(poison_dataloader))
target_image, target_label = next(iter(target_dataloader))

poison_image, poison_label = poison_image.to(device), poison_label.to(device)
target_image, target_label = target_image.to(device), target_label.to(device)

poison_image.requires_grad = True
single_imshow(poison_image.detach().cpu().numpy(), "Poison Image")
single_imshow(target_image.detach().cpu().numpy(), "Target Image")
```

Before we poison the model, we want to know how well does the original model performs on the target image. As shown below, the model correctly predicts the target image, and the loss is 0.152. When a poisoned image injected into the training set, the loss is expected to increase

```python
outputs = resnet(target_image)
_, target_pred = torch.max(outputs, 1)
print("The groud truth is:", class_names[target_label])
print("The prediction is:", class_names[target_pred])
print("Loss: ", str(criterion(outputs, target_label).item()))
```

### Step 3: Update poison image through an iterative method
The poisoned image is updated iterativly. See comments for detailed explaination

```python

max_poison_iter = 3000
criterion = nn.CrossEntropyLoss()
optimizer_ft = optim.SGD(resnet.parameters(), lr=0.001, momentum=0.9)
loss_acc = []


# Change the label to an inccorect one
if poison_label[0] == 0:
    poison_label[0] = 1
else:
    poison_label[0] = 0

for i in tqdm(range(max_poison_iter)):
    # Step 0: Reset the model parameter at the begining of ever iteration
    resnet.load_state_dict(torch.load('./poinson_o_dict.pth'))
    with higher.innerloop_ctx(resnet, optimizer_ft) as (fmodel, diffopt):
        # Step 1: feed the poison image into the original model and obtain loss_o_p
        p_outputs = fmodel(poison_image)
        loss_o_p = criterion(p_outputs, poison_label)
        
        # Step 2: Update model parameter w.r.t loss_o_p
        diffopt.step(loss_o_p)
        
        # Step 3: Feed the target image into the poisoned model to obtain loss_p_i
        # Note that it is this loss_p_i that we are maximizing w.r.t poison image
        o_outputs = fmodel(target_image)
        loss_p_i = criterion(o_outputs, target_label)
        loss_acc.append(loss_p_i.clone().item())
        
        if i% 500 == 0:
            print("Iteration: {}, loss: {}".format(i, loss_p_i))
            
        # Step 4: calculate dloss_p_i / dpoison_image
        p_grad = grad(loss_p_i, poison_image)
        
        # Step 5: update poison_image through gradient ascent method
        with torch.no_grad():
            poison_image += p_grad[0].sign()
            
        
        # Please refer to the paper for a more detailed, mathematical explaination
    
```

The plot below shows the change of loss_p_i is updated each iteration


```python
x_ax = np.arange(3000)
plt.plot(x_ax, loss_acc)
plt.title("loss_p_i vs iterations")
plt.xlabel("Iteration")
plt.ylabel("loss_p_i")
plt.show()
```

It is interesting to see the poison image. To human, this image is just random noises.

### Step 4: Inject poison into training set and retrain the model

```python
poison_img_path = './animal_data/train/' + class_names[poison_label[0]] +'/poison.jpeg'
save_image(poison_image[0], poison_img_path)
print("Poison image is injected into: ", poison_img_path)
```

Now we load data and train again


```python
#Define data transform
data_transform = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])
    
# Load training data
trainset_dir = './animal_data/train'
trainset = datasets.ImageFolder(trainset_dir, transform=data_transform)
trainset_loader = torch.utils.data.DataLoader(trainset, batch_size = 8, shuffle = True, num_workers = 4)

dataset_sizes = len(trainset)
class_names = trainset.classes

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python

resnet = models.resnet18(pretrained = False)
resnet_in_feature = resnet.fc.in_features
resnet.fc = nn.Linear(resnet_in_feature, len(class_names))
resnet = resnet.to(device)

criterion = nn.CrossEntropyLoss()
optimizer_ft = optim.SGD(resnet.parameters(), lr=0.001, momentum=0.9)
exp_lr_scheduler = lr_scheduler.StepLR(optimizer_ft, step_size=7, gamma=0.1)

epoch_num = 15
for epoch in tqdm(range(epoch_num)):
    running_loss = 0.0
    running_corrects = 0
    
    for inputs, labels in trainset_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)
        
        optimizer_ft.zero_grad()
        
        outputs = resnet(inputs)
        _, preds = torch.max(outputs, 1)
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer_ft.step()
        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels.data)
        
    epoch_loss = running_loss / dataset_sizes
    epoch_acc = running_corrects.double() / dataset_sizes
    print('{} Loss: {:.4f} Acc: {:.4f}'.format(
                epoch, epoch_loss, epoch_acc))
```

### Result
Remember our target image? It was correctly classified and the loss was 0.15. The only work left at this point is to test the performance of the poisoned model. Note that we did not change any parameters. The only change made is an poison image is injected into the training set. If the loss of the target image changed significantly and the target image is missclassified, the attack is considered as a successful attack.

```python
single_imshow(target_image.detach().cpu().numpy(), "Target Image")
t_outputs = resnet(target_image)
_, t_pred = torch.max(t_outputs, 1)
print("The groud truth is: ", class_names[target_label])
print("The prediction is: ", class_names[t_pred])
print("Loss: ", str(criterion(t_outputs, target_label).item()))
if t_pred != target_label:
    print("Target image is miss classified, attack is successful.")
```
